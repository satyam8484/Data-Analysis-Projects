### Data Imputation and Manipulation
#### Fill the null values with appropriate values using aggregate functions such as mean, median or mode

```python
google_data['Rating'].fillna(google_data['Rating'].median(),inplace= True)
```

```python
google_data.isnull().sum()
```

```python
print(google_data['Type'].mode())
print(google_data['Current Ver'].mode())
print(google_data['Android Ver'].mode())
```

```python
google_data['Type'].fillna(str(google_data['Type'].mode().values[0]),inplace=True)
google_data['Current Ver'].fillna(str(google_data['Current Ver'].mode().values[0]),inplace=True)
google_data['Android Ver'].fillna(str(google_data['Android Ver'].mode().values[0]),inplace=True)
```

```python
google_data.isnull().sum()
```

```python
google_data['Price'] = google_data['Price'].apply(lambda x: str(x).replace("$","") if "$" in str(x) else str(x))
google_data['Price'] = google_data['Price'].apply(lambda x:float(x))
google_data['Price'].unique()
```

```python
google_data['Reviews'] = pd.to_numeric(google_data['Reviews'],errors='coerce')
google_data['Reviews'].head()
```

```python
google_data['Installs'] = google_data['Installs'].apply(lambda x:str(x).replace("+","").replace(",","") if "+" in str(x) else str(x))
google_data['Installs'] = google_data['Installs'].apply(lambda x:float(x))
google_data['Installs'].unique()
```

```python
grp = google_data.groupby('Category')
x = grp['Rating'].agg(np.mean)
y = grp['Reviews'].agg(np.mean)
z = grp['Price'].agg(np.mean)

print(x)
print(y)
print(z)
```

```python
x_pos = np.arange(len(x))
x_pos
```

```python
plt.figure(figsize=(12,5))
plt.title("Category wise ratings")
plt.ylabel("Ratings -->")
plt.xlabel("categories -->")
plt.xticks(x_pos,x.index,rotation=90)
plt.plot(x_pos,x.values,'ro',color='r')
```

```python
plt.figure(figsize=(16,5))
plt.title('Categories wise Review')
plt.xlabel('categorise -->')
plt.ylabel('reviews --> ')
y_pos = np.arange(len(y))
plt.xticks(y_pos,x.index,rotation=90)
plt.plot(y_pos,y.values,'r--',color='b')
```

```python
plt.figure(figsize=(16,5))
plt.title('Categories wise Price')
plt.xlabel('categorise -->')
plt.ylabel('price --> ')
z_pos = np.arange(len(y))
plt.xticks(z_pos,z.index,rotation=90)
plt.plot(z_pos,z.values,'bs',color='g')
plt.grid()
```
